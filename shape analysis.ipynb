```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

```python
def plot(data):
    mu = np.mean(data)
    sigma = np.std(data)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    y=stats.norm.pdf(x, mu, sigma)
    plt.plot(x, y)
    plt.hist(data,bins = 'auto', density = 'True')# geef commando voor plotten
    plt.xlabel("circularity")
    plt.ylabel("frequency")
    plt.show()
    print(stats.t.interval(confidence=0.90, df=len(data)-1,
              loc=np.mean(data),
              scale=stats.sem(data)))
```
