```python
def plot(data):
    mu = np.mean(data)
    sigma = np.std(data)
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
    y=stats.norm.pdf(x, mu, sigma)
    plt.plot(x, y, label="distribution")
    plt.hist(data,bins = 'auto', density = 'True', label="histogram")# geef commando voor plotten
    plt.axvline(mu, label="mean", color="red")
    plt.legend()
    plt.show()
    print("The mean value is", mu)
    
```
