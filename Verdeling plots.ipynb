```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import lognorm
from scipy import stats
```

```python
#pel de pinda is oke 
def plot(data):
    data = data*302
    # Fit the data to a log-normal distribution
    shape, loc, scale = lognorm.fit(data)
    # Generate log-normal distributed data for plotting
    x = np.linspace(min(data), max(data), 1000)
    pdf = lognorm.pdf(x, shape, loc, scale)
    hist, bins, _ = plt.hist(data, bins=15)
    logbins = np.logspace(np.log10(bins[0]),np.log10(bins[-1]),len(bins))
    plt.show()

    # Plot the histogram of the data
    plt.hist(data, bins=logbins, label='Data',density = True)
#     plt.xscale('log')

    # Plot the fitted log-normal distribution
    plt.plot(x, pdf, 'r-', lw=2, label='Log-Normal Fit')
    # Customize the plot
    plt.xlabel('area(um)')
    plt.ylabel('Probability Density')
    plt.title('Particle size distribution')
    plt.legend()


    # Show the plot
    plt.show()
    print(stats.t.interval(confidence=0.90, df=len(data)-1,
          loc=np.mean(data),
          scale=stats.sem(data)))
```
